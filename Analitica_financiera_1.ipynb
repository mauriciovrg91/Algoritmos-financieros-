```python
#API DATOS
import yfinance as yfin

#manejo de tablas de datos
import pandas as pd

#Manejo de archivos
from google.colab import files

# Graficos
import matplotlib.pyplot as plt
```

```python
ticker = "AAPL"
df = yfin.download(ticker, start="2019-01-01", end="2021-01-01")
df
```

```python
# Guardar en .csv
df.to_csv("AAPL.CSV")

# Descargar .csv
files.download("AAPL.CSV")
```

```python
# Cargar Archivo
files.upload()
```

```python
# Data frame 2
df2 = pd.read_csv("AAPL.CSV", index_col='Date')
df2.index = pd.to_datetime(df2.index)
df2
```

```python
# estadistica descriptiva
df2.describe()
```

```python
# Grafico de los precios de cierre
plt.figure(figsize=(15,8))
plt.plot(df2['Date'])
plt.title('Precio de cierre de AAPL')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')
plt.show()
```
